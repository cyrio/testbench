## Essai 1er DF

### 1. Distribution des valeurs des capteurs

```python
fig=plt.figure(figsize=(30,30))
plt.subplot(3,3,1)
df.boxplot(column='SENSOR_1')
plt.subplot(3,3,2)
df.boxplot(column='SENSOR_2')
plt.subplot(3,3,3)
df.boxplot(column='SENSOR_3')
plt.subplot(3,3,4)
df.boxplot(column='SENSOR_4')
plt.subplot(3,3,5)
df.boxplot(column='SENSOR_5')
```

```python
df.plot(y=['SENSOR_1', 'SENSOR_2','SENSOR_3','SENSOR_4','SENSOR_5'], subplots=True, layout=(4,4), figsize=(15,15))
```

### 2. Distribution des valeurs de chargement 

```python
fig=plt.figure(figsize=(30,30))
df.plot(y=['PR_1', 'PR_2','PR_3','PR_4','TEMP_1','TEMP_2','BRAKE_1'], subplots=True, layout=(4,4), figsize=(15,15))
```
